```python
import pandas as pd
import os
from pathlib import Path
import cv2
import numpy as np 
```

```python
def load_and_filter(
    path,
    key,
    values,
    output_path=None
):
    path = Path(path)

    if path.suffix == ".csv":
        df = pd.read_csv(path)
    elif path.suffix in [".xlsx", ".xls", ".ods"]:
        df = pd.read_excel(path)
    else:
        raise ValueError(f"Format non supporté : {path.suffix}")

    if not isinstance(values, (list, set, tuple)):
        values = [values]

    # --- filtrage
    df_filtered = df[~df[key].isin(values)].reset_index(drop=True)
    print(f"Filtrage effectué : {len(df) - len(df_filtered)} lignes supprimées.")

    # --- sauvegarde optionnelle
    if output_path is not None:
        output_path = Path(output_path)
        if output_path.suffix == ".csv":
            df_filtered.to_csv(output_path, index=False)
        else:
            df_filtered.to_excel(output_path, index=False)

    return df_filtered
```

```python
DIR = "path/to/annotation/file"
df = load_and_filter(
    os.path.join(DIR,"annotations_trills.csv" ),
    key="annotator",
    values=["test"],
    output_path=os.path.join(DIR,"annotations_trills_clean.csv")
)
```
